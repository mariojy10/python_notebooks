# Conditionals
Conditionals (expression and statements) are used to determine program flow (orderof  execution of the statements in program)

## if Statement
General format:
```
if test:
    statement
```
`test` is an expression evaluated in Boolean context  
`statement` is a valid Python statement, which must be indented, if `test` evaluated as True statement is executed


````python
if True:
    print(1)
````

````python
grade = 70
minimum_score = 75

if grade < minimum_score:
    print('You dont pass the test')
````

````python
from math import sqrt
point  = {'x':3, 'y': 4}

if 'x' in point:
    print('x = ',point['x'])
    if 'y' in point:
        print('y = ',point['y'])
        r = sqrt((point['x']**2 + point['y']**2))
        print('r = ',int(r))
        
````

## else Clause
General format:
```
if test:
    statement1
else:
    statement2
```
If `test` evaluated as True then `statement1` is executed and `statement2` is skipped, else if `test` evaluated as False then `statement1` is skipped and `statement2` is executed

````python
x = 0
y = 10

if x > y:
    print(x)
else:
    print(y)
````

````python
grade = 100
minimum_score = 75

if grade > minimum_score:
    print('You pass the test')
    if grade == 100:
        print('With perfect score')
else:
    print('You dont pass the test')
````

## elif Clause
General format:
```
if test1:
    statement1
elif test2:
    statement2
elif test3:
    statement3
else:
    statement4
```
To facilitate branching execution based on several alternatives use one or more `elif` clauses. Python evaluates each test in turn and executes the statement to the first that is true

````python
grade = 85
if grade >= 90:         # 1st test evaluated as False
    print("You got A")  
elif grade >= 80:       # 2nd test evaluated as True, execute 2nd statement, exit branches
    print("You got B.")
elif grade >= 70:
    print("You got C.")
else:
    print("You did not pass the class")
````

## One-Line if Statement

```
if test: statement
```

````python
x = 10

if x > 0 and x % 2 == 0: print(x)
````

```
if test: statement1; statement2; statement3
```

````python
l = [1,2,3,4,5]

if 2 in l: print('2'); print('index: ',l.index(2)); print('count: ',l.count(2))
````

## Conditional Expressions (Python's Ternary Operator)
General Format:
```
statement1 if test else statement2
```
First `test` is evaluated, if evaluated as True statement1 is executed else statement2 is executed

````python
age = 50

old = True if age > 45 else False

old
````

````python
x = 20
y = 40

z = x + y if x < y else x + 2

z
````
